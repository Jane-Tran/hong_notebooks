# Phương trình bậc nhất, 1 ẩn

$$A(x) = B(x)$$

- Phương trình tương đương khi có cùng một tập nghiệm (solution).
- Phương trình tích:
$$A(x).B(x) = 0 \iff \left[ \begin{array}{l}A(x) = 0\\B(x) = 0\end{array} \right.$$

- Phương trình chứa ẩn ở mẫu $\frac{numerator}{denominator}$
$$\frac{{{A_1}(x)}}{{{B_1}(x)}} + \frac{{{A_2}(x)}}{{{B_2}(x)}} + ... + \frac{{{A_n}(x)}}{{{B_n}(x)}} = 0$$,

Điều kiện: $$\left\{ \begin{array}{l}{B_1}(x) \ne 0\\{B_2}(x) \ne 0\\.........\\{B_n}(x) \ne 0\end{array} \right.$$

- Giải toán bằng cách lập phương trình

# Bất phương trình bậc nhất 1 ẩn (disequation)

$$a \ge b \mid a,b \in \mathbb{R}$$

**Tổng**

if $a > b$ then $a + c > b + c$

if $a \ge b$ then $a + c \ge b + c$

**Tích**

if $a > b$ then $a.c > b.c \mid c > 0$

if $a \ge b$ then $a.c \ge b.c \mid c > 0$

if $a > b$ then $a.c < b.c \mid c < 0$

if $a \ge b$ then $a.c \le b.c \mid c < 0$

**Bắc cầu (bridging)**

if $a > b, b > c$ then $a > c$

- Bất phương trình 1 ẩn
$$A(x) > B(x), A(x) < B(x)$$

$$A(x) \ge B(x), A(x) \le B(x)$$

- Bất phương trình tương đương khi có cùng một tập nghiệm.
- Biến đổi bất phương trình:
$$a + b < c \iff a + b - c < 0$$

- Bất phương trình bậc nhất 1 ẩn:
$$ax + b > 0, ax + b < 0 (a \ne 0)$$

$$ax + b \ge 0, ax + b \le 0 (a \ne 0)$$

- Phương trình chứa dấu $|\bullet|$

$$|a| = \left\{ \begin{array}{l}a\,\,\,neu\,\,\,a \ge 0\\ - a\,\,neu\,\,a\,\, < \,\,0\end{array} \right.$$

$$|f(x)| = \left\{ \begin{array}{l}f(x)\,\,\,neu\,\,f(x)\, \ge 0\\ - f(x)\,\,neu\,\,f(x)\, < \,0\end{array} \right.$$

- Các trường hợp
$$\left| {f(x)} \right| = k \Leftrightarrow \,\left[ \begin{array}{l}f(x) = k\\f(x) =  - k\end{array} \right.$$

$$|f(x)| = |g(x)| \Leftrightarrow \left[ \begin{array}{l}f(x) = g(x)\\f(x) =  - g(x)\end{array} \right.$$

$$|f(x)| = g(x) \Leftrightarrow \left[ \begin{array}{l}f(x) = g(x)\\f(x) =  - g(x)\end{array} \right. where \ g(x) \ge 0$$
